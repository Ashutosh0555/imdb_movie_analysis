## Importing pandas

```python
import pandas as pd
```

```python
mv = pd.read_csv('imdb.csv')
print(mv.columns)
```

```python
print(mv['Series_Title'])
```

#### Creating a new column of popularity score 

```python
mv['Popularity_Score']= mv['IMDB_Rating']*mv['No_of_Votes']
```

#### Sorting in decending order on basis of popularity score

```python
mv =mv.sort_values('Popularity_Score', ascending = False)
```

#### Creating a new file of only movies rated above 8.5 IMDB

```python
new_mv= mv.loc[mv['IMDB_Rating']>8.5]
```

##### Creating a  file of top 20 movies of that new file

```python
top_10= new_mv.head(20)
top_10
```

```python
top_10.to_csv("top_movies.csv", index=False)
```
